```python
import pandas as pd
```

```python
# read date from xlsx
df = pd.read_excel('Merge_excel.xlsx')
df.info()
```

```python
df.head(1)
```

```python
# drop cloumn: reference, image_url, 網址來源 (useless)
#               症状对应描述1~10, 针灸穴位 (NULL column)
try:
    df = df.drop(columns=['reference','image_url','症状对应描述1','症状对应描述2',
                    '症状对应描述3','症状对应描述4','症状对应描述5','症状对应描述6',
                    '症状对应描述7','症状对应描述8','症状对应描述9','症状对应描述10',
                    '针灸穴位' ,'網址來源'])
except:
    df.head()
df.head(1)
```

```python
# replace NaN to None
df = df.where(pd.notnull(df), None)
df.head()
```

```python
df.loc[4,'acupoint_ID']
```

```python
# same disease_name => same 症状s
# different only from acupoints

# first, combine 症状s (symptoms)
df.reset_index()
df['symptoms'] = [[None]] * len(df)
for index, row in df.iterrows():
    df.loc[index,'symptoms'] = []
    for column in ['症状1','症状2','症状3','症状4','症状5','症状6','症状7','症状8','症状9','症状10']:
        if df.loc[index,column] != None:
            df.loc[index,'symptoms'].append(df.loc[index,column])
# then, delete 症状1~10
try:
    df = df.drop(columns=['症状1','症状2','症状3','症状4','症状5','症状6','症状7','症状8','症状9','症状10'])
except:
    df.head()
df.head(6)
```

```python
# next, combine acupoints from same disease
df_uniq_disease = df.copy()
try:
    df_uniq_disease = df_uniq_disease.drop(columns=['acupoint_name'])
except:
    df_uniq_disease.head()
df_uniq_disease = df_uniq_disease.drop_duplicates(subset = 'disease_name')
df_uniq_disease
# then we can get unique disease_name list

# second, create new columns called 'acupoints'
df['acupoints'] = [[None]] * len(df)
for index, row in df.iterrows():
    df.loc[index,'acupoints'] = []
for index, row in df_uniq_disease.iterrows():
    disease_name = df.loc[index,'disease_name']
    # seek down until meet different disease_name
    tmp_index = index
    while True:
        try:
            df.loc[tmp_index,'disease_name']
        except:
            break
        
        if df.loc[tmp_index,'disease_name'] != disease_name :
            break
        else:
            # avoid duplicate acupoints
            if df.loc[tmp_index,'acupoint_name'] not in df.loc[index,'acupoints']:
                df.loc[index,'acupoints'].append(df.loc[tmp_index,'acupoint_name'])
            tmp_index = tmp_index+1
df
```

```python
# drop column acupoint_name, acupoint_ID and drop duplicate from disease_name:
try:
    df = df.drop(columns=['acupoint_ID','acupoint_name'])
except:
    df.head()
df = df.drop_duplicates(subset = 'disease_name')
df
```

```python
# save as csv
df.to_csv('disease_to_acu_and_symptom.csv', index=False,encoding='utf_8_sig')
```

```python
# caculate data info
all_acupoint_list = []
all_disease_list = []
for index, row in df.iterrows():
    if df.loc[index,'disease_name'] not in all_disease_list:
        all_disease_list.append(df.loc[index,'disease_name'])
    for acupoint in df.loc[index,'acupoints']:
        if acupoint not in all_acupoint_list:
            all_acupoint_list.append(acupoint)

print('# disease: ', len(all_disease_list))
print('# acupoint: ', len(all_acupoint_list))
```
