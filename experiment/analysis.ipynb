```python
import pandas as pd
import ast
```

```python
df = pd.read_csv('disease_to_acu_and_symptom_ZH.csv')
df.head(3)
```

```python
df['symptoms'] = df['symptoms'].apply(lambda x: ast.literal_eval(x))
df['acupoints'] = df['acupoints'].apply(lambda x: ast.literal_eval(x))
print(type(df.loc[0,'symptoms']), type(df.loc[0,'acupoints']))
```

```python
df
```

```python
all_acupoint_list = []
all_disease_list = []
all_symptom_list = []
no_symptoms = []
itself_is_symptom = []

all_acu= 0
for index, row in df.iterrows():
    # disease
    if df.loc[index,'disease_name'] not in all_disease_list:
        all_disease_list.append(df.loc[index,'disease_name'])
    # acupoint
    for acupoint in df.loc[index,'acupoints']:
        all_acu+=1
        if acupoint not in all_acupoint_list:
            all_acupoint_list.append(acupoint)
    # check if this disease has no symptom
    if len(df.loc[index,'symptoms']) == 0:
        no_symptoms.append(df.loc[index,'disease_name'])
    # check if this disease is symptom
    if len(df.loc[index,'symptoms']) == 1 and df.loc[index,'symptoms'][0] == df.loc[index,'disease_name']:
        itself_is_symptom.append(df.loc[index,'disease_name'])
    # symptom
    for symptom in df.loc[index,'symptoms']:
        if symptom not in all_symptom_list:
            all_symptom_list.append(symptom)
print(all_acu/len(all_disease_list))
```
